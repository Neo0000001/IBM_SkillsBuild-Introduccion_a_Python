## El bucle for

## El bucle while

```python
# WHILE

# Imprime edad cuando el contador llegue a 18
edad = 0

while edad < 18:
    edad = edad + 1
    print("Tienes " + str(edad))

# Pregunta la edad mientras sea negativa
edad = int(input("Introduce edad: "))

while edad < 0:
    print("Edad incorrecta")
    edad = int(input("Introduce edad: "))

print("tu edad es: " + str(edad))

# Calcula la raíz cuadrada de un número. Tenemos tres intentos y el número no puede ser negativo.
import math

intentos = 0
num = int(input("Introduce número: "))

while num < 0:
    intentos = intentos + 1
    print("Incorrecto")
    num = int(input("Introduce número: "))
    if intentos == 2:
        print("Agotaste intentos")
        break

if intentos < 2:
    solucion = math.sqrt(num)
    print("La raíz cuadrada es " + str(solucion))

# Bucle while anidado
x = 1

while x < 6:
    y = 1

    while y < 6:
        print("*", end=" ")
        y += 1

    print("")
    x += 1

# Con la declaración break podemos detener el ciclo incluso si la condición while es verdadera.
i = 1

while i < 6:
    print(i)

    if i == 3:
        break

    i += 1

# Con la declaración continue podemos detener la iteración actual y continuar con la siguiente.
i = 0

while i < 6:
    i += 1

    if i == 3:
        continue
    
    print(i)

# Con la declaración else podemos ejecutar un bloque de código una vez cuando la condición ya no sea verdadera.
i = 1

while i < 6:
    print(i)
    i += 1
else:
    print("i no es menor que 6")

# Aunque el bucle while requiere una variable de indexado ready. También podemos hacerlo con listas.
# En este caso la variable de indexado es lista.
i = 0

lista = ["primavera", "verano", "otoño", "invierno"]

while i < len(lista):
    print(lista[i])
    i += 1
```

## Listas

```python
# LISTAS
# Creamos una lista de 100 elementos del 1 al 100
lista = []

for i in range(1, 101):
    lista.append(i)

print(lista)

# Cargamos la lista con valores random
import random

for i in range(1, 101):
    lista.append(random.randint(1, 100))
    
print(lista)

# Realiza la media de los 100 valores contenidos en la lista
suma = 0

for valor in lista:
    suma += valor

media = suma / len(lista)
print(f"La media de los 100 valores es: {media}")

# Devuelve la posición donde se encuentra la primera ocurrencia del número 50 en la lista
posicion = lista.index(50)
print(f"La primera ocurrencia de 50 en la lista se encuentra en la posición: {posicion}")

# Ordena la lista de menor a mayor
lista.sort()
print("Lista ordenada de menor a mayor:")
print(lista)

# Invierte el orden de la lista
lista.reverse()
print("Lista invertida:")
print(lista)

# Imprime la lista con los valores únicos
lista_unica = list(set(lista))
print("Lista con valores únicos:")
print(lista_unica)
```

## Funciones

```python
# FUNCIONES

# Definimos una función que suma dos números
def suma(a, b):
    return a + b


# Llamamos a la función con dos números
resultado = suma(3, 5)
print(f"La suma de 3 y 5 es: {resultado}")


# Definimos una función que calcula la media de una lista de números
def media(lista):
    return sum(lista) / len(lista)


# Llamamos a la función con una lista de números
lista = [1, 2, 3, 4, 5]
resultado = media(lista)
print(f"La media de la lista es: {resultado}")


# Definimos una función que devuelve la lista con los valores únicos
def valores_unicos(lista):
    return list(set(lista))


# Llamamos a la función con una lista de números
lista = [1, 2, 2, 3, 3, 4, 5]
resultado = valores_unicos(lista)
print(f"La lista con valores únicos es: {resultado}")


# Definimos una función que calcula el factorial de un número
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)


# Llamamos a la función con un número
numero = 5
resultado = factorial(numero)
print(f"El factorial de {numero} es: {resultado}")


# Definimos una función que verifica si un número es primo
def es_primo(n):
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
        
    return True


# Llamamos a la función con un número
numero = 7
resultado = es_primo(numero)
print(f"¿El número {numero} es primo? {'Sí' si resultado else 'No'}")


# Definimos una función que devuelve los números primos en un rango dado
def primos_en_rango(inicio, fin):
    primos = []

    for num in range(inicio, fin + 1):
        if es_primo(num):
            primos.append(num)

    return primos


# Llamamos a la función con un rango
inicio = 10
fin = 20
resultado = primos_en_rango(inicio, fin)
print(f"Los números primos entre {inicio} y {fin} son: {resultado}")


# Definimos una función que encuentra el máximo común divisor (MCD) de dos números
def mcd(a, b):
    while b:
        a, b = b, a % b
    return a


# Llamamos a la función con dos números
num1 = 48
num2 = 18
resultado = mcd(num1, num2)
print(f"El MCD de {num1} y {num2} es: {resultado}")


# Definimos una función que encuentra el mínimo común múltiplo (MCM) de dos números
def mcm(a, b):
    return abs(a * b) // mcd(a, b)


# Llamamos a la función con dos números
num1 = 12
num2 = 15
resultado = mcm(num1, num2)
print(f"El MCM de {num1} y {num2} es: {resultado}")


# Definimos una función que invierte una cadena de texto
def invertir_cadena(cadena):
    return cadena[::-1]


# Llamamos a la función con una cadena de texto
texto = "Hola, mundo"
resultado = invertir_cadena(texto)
print(f"La cadena invertida de '{texto}' es: '{resultado}'")


# Definimos una función que cuenta el número de vocales en una cadena de texto
def contar_vocales(cadena):
    vocales = "aeiouAEIOU"
    contador = 0

    for char in cadena:
        if char in vocales:
            contador += 1
            
    return contador


# Llamamos a la función con una cadena de texto
texto = "Hola, mundo"
resultado = contar_vocales(texto)
print(f"El número de vocales en '{texto}' es: {resultado}")
```

## Clases y Objetos

```python
# CLASES Y OBJETOS

# Definimos una clase que representa un coche
class Coche:
    def __init__(self, marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año
    
    def mostrar_informacion(self):
        print(f"Marca: {self.marca}, Modelo: {self.modelo}, Año: {self.año}")


# Creamos un objeto de la clase Coche
mi_coche = Coche("Toyota", "Corolla", 2020)

# Llamamos al método mostrar_informacion del objeto
mi_coche.mostrar_informacion()

# Definimos una clase que representa un círculo
import math


class Circulo:
    def __init__(self, radio):
        self.radio = radio
    
    def calcular_area(self):
        return math.pi * (self.radio ** 2)
    
    def calcular_perimetro(self):
        return 2 * math.pi * self.radio


# Creamos un objeto de la clase Circulo
mi_circulo = Circulo(5)

# Llamamos a los métodos del objeto
area = mi_circulo.calcular_area()
perimetro = mi_circulo.calcular_perimetro()
print(f"Área del círculo: {area}")
print(f"Perímetro del círculo: {perimetro}")


# Definimos una clase que representa un estudiante
class Estudiante:
    def __init__(self, nombre, edad, curso):
        self.nombre = nombre
        self.edad = edad
        self.curso = curso
    
    def mostrar_informacion(self):
        print(f"Nombre: {self.nombre}, Edad: {self.edad}, Curso: {self.curso}")


# Creamos un objeto de la clase Estudiante
mi_estudiante = Estudiante("Juan", 20, "Ingeniería Informática")

# Llamamos al método mostrar_informacion del objeto
mi_estudiante.mostrar_informacion()
```

## Manejo de archivos


```python
# MANEJO DE ARCHIVOS

# Escribimos en un archivo
with open("archivo.txt", "w") as archivo:
    archivo.write("Hola, mundo\n")
    archivo.write("Esta es una segunda línea\n")

# Leemos el contenido de un archivo
with open("archivo.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)

# Leemos el archivo línea por línea
with open("archivo.txt", "r") as archivo:
    for linea in archivo:
        print(linea, end="")

# Añadimos contenido a un archivo
with open("archivo.txt", "a") as archivo:
    archivo.write("Esta es una línea adicional\n")

# Leemos el contenido actualizado del archivo
with open("archivo.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)

# Leemos todas las líneas del archivo en una lista
with open("archivo.txt", "r") as archivo:
    lineas = archivo.readlines()
    print(lineas)
```

## Manejo de excepciones


```python
# MANEJO DE EXCEPCIONES

# Ejemplo de manejo de excepciones con división por cero
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("Error: División por cero")
else:
    print(f"El resultado es: {resultado}")
finally:
    print("La operación ha finalizado")

# Ejemplo de manejo de excepciones con acceso a archivo
try:
    with open("archivo_inexistente.txt", "r") as archivo:
        contenido = archivo.read()
except FileNotFoundError:
    print("Error: El archivo no existe")
else:
    print(contenido)
finally:
    print("La operación de lectura ha finalizado")

# Ejemplo de manejo de múltiples excepciones
try:
    resultado = 10 / 0
    with open("archivo_inexistente.txt", "r") as archivo:
        contenido = archivo.read()
except ZeroDivisionError:
    print("Error: División por cero")
except FileNotFoundError:
    print("Error: El archivo no existe")
finally:
    print("La operación ha finalizado")
```

## Programación orientada a objetos avanzada


```python
# PROGRAMACIÓN ORIENTADA A OBJETOS AVANZADA

# Definimos una clase base y una clase derivada
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    
    def mostrar_informacion(self):
        print(f"Nombre: {self.nombre}, Edad: {self.edad}")

class Estudiante(Persona):
    def __init__(self, nombre, edad, curso):
        super().__init__(nombre, edad)
        self.curso = curso
    
    def mostrar_informacion(self):
        super().mostrar_informacion()
        print(f"Curso: {self.curso}")

# Creamos un objeto de la clase derivada y llamamos a sus métodos
mi_estudiante = Estudiante("Juan", 20, "Ingeniería Informática")
mi_estudiante.mostrar_informacion()

# Definimos una clase abstracta
from abc import ABC, abstractmethod

class Figura(ABC):
    @abstractmethod
    def calcular_area(self):
        pass


class Rectangulo(Figura):
    def __init__(self, ancho, alto):
        self.ancho = ancho
        self.alto = alto
    
    def calcular_area(self):
        return self.ancho * self.alto


# Creamos un objeto de la clase derivada y llamamos a sus métodos
mi_rectangulo = Rectangulo(10, 5)
area = mi_rectangulo.calcular_area()
print(f"El área del rectángulo es: {area}")
```

## Decoradores

```python
# DECORADORES

# Definimos un decorador que mide el tiempo de ejecución de una función
import time

def medir_tiempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fin = time.time()
        print(f"Tiempo de ejecución: {fin - inicio} segundos")
        return resultado
    
    return wrapper


# Aplicamos el decorador a una función
@medir_tiempo
def contar_hasta(numero):
    contador = 0
    while contador < numero:
        contador += 1

    return contador


# Llamamos a la función decorada
resultado = contar_hasta(1000000)
print(f"Contador final: {resultado}")


# Definimos un decorador que verifica permisos
def verificar_permiso(func):
    def wrapper(*args, **kwargs):
        if args[0].permiso:
            return func(*args, **kwargs)
        else:
            print("Permiso denegado")
            
    return wrapper


# Definimos una clase que utiliza el decorador
class Usuario:
    def __init__(self, nombre, permiso):
        self.nombre = nombre
        self.permiso = permiso

    @verificar_permiso
    def acceder_recurso(self):
        print("Recurso accesado")


# Creamos objetos de la clase Usuario y llamamos al método decorado
usuario_con_permiso = Usuario("Alice", True)
usuario_sin_permiso = Usuario("Bob", False)

usuario_con_permiso.acceder_recurso()
usuario_sin_permiso.acceder_recurso()
```
