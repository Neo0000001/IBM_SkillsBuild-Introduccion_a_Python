## Introducción

Cuando un sistema se prueba exhaustivamente por unidad, las pruebas de integración se vuelven mucho más fáciles, ya que muchos de los errores en los componentes individuales ya se habrán encontrado y solucionado.

Hoy en día, los proyectos de software son más complejos que nunca. La industria del software crece a una velocidad vertiginosa y los desarrolladores deben mantenerse al tanto de las novedades y oportunidades en el mundo del software para utilizar la última tecnología de la mejor manera posible.

En el desarrollo de software, la calidad es esencial para el éxito. Para garantizar este éxito, la función de testing es vital. Por lo tanto, es necesario realizar pruebas tanto de integración como unitarias desde el principio del desarrollo. La combinación de ambas garantiza que el código sea de buena calidad y realmente solucione los problemas para los cuales fue desarrollado.

A medida que la escala de una base de código aumenta, tanto las pruebas unitarias como las de integración permiten a los desarrolladores identificar rápidamente cambios importantes en su código. Estos cambios a menudo no son intencionados y pueden no ser conocidos hasta más adelante en el ciclo de desarrollo, posiblemente cuando un usuario final descubra el problema mientras usa el software. Las pruebas automatizadas de unidad e integración aumentan en gran medida la probabilidad de que los errores se encuentren lo antes posible durante el desarrollo para que se puedan solucionar de inmediato.

La prueba unitaria en Python es un marco integrado para probar el código Python. Tiene un corredor de pruebas que nos permite ejecutar las pruebas sin mucho esfuerzo. Podemos usar el módulo de prueba unitaria para probar sin utilizar módulos de terceros.

Mientras que las pruebas unitarias corren pruebas sobre partes del código de forma independiente, las pruebas de integración se hacen necesarias cuando queremos probar de manera sistémica un proyecto como un todo. Ejercitan dos o más partes de una aplicación a la vez, incluidas las interacciones entre ellas, para determinar si funcionan según lo previsto. Este tipo de prueba identifica defectos en las interfaces entre partes dispares de una base de código a medida que se invocan y pasan datos entre sí.

Estos dos enfoques deben utilizarse juntos en lugar de enfocarse solo en uno sobre el otro.


## Pruebas Unitarias

Pruebas Unitarias
El módulo Unittest de la librería estándar (stdlib) de Python es el encargado de llevar a cabo las pruebas unitarias si no queremos acudir a librerías externas. Se le conoce también como PyUnit, donde el sufijo "Unit" hace referencia a un conjunto de parámetros construidos exprofeso para ejecutar pruebas unitarias en diferentes lenguajes de programación, y que se conocen como XUnit, donde "X" es el prefijo que apunta al lenguaje de programación en concreto. En nuestro caso, "Py", el prefijo, señala a Python.

Este módulo Unittest cuenta con una serie de herramientas que nos permitirán probar nuestro programa o aplicación dentro del mismo código. Si partimos de la base de que la prueba unitaria es algo así como la "unidad básica" de prueba o testeo en un programa/aplicación, empecemos por ella para ir poco a poco abordando aspectos un poquito más complejos más adelante.

Las pruebas unitarias son un primer filtro para la detección de fallos en los sistemas. Sin embargo, estas no identifican su relación con otras interfaces. De ahí la importancia de implementar pruebas de integración de software.

Entre los problemas más comunes que identifican este tipo de pruebas se encuentran la pérdida de conectividad, el formateo de datos y las respuestas inesperadas.

## Pruebas de integración

Las pruebas de integración de software son la herramienta que conjunta cada uno de los módulos de un sistema para comprobar su funcionamiento entre sí.
Este tipo de test se realizan en las primeras etapas, después de las pruebas unitarias, en las que se analiza un fragmento del código fuente.

Una vez analizadas las unidades por separado, se debe verificar que los módulos no interfieren con el resto de las funciones.

Imaginemos que una aplicación de correo electrónico está dividida en tres unidades —página de inicio, bandeja de entrada y papelera—. Al ejecutar una prueba de integración el desarrollador debe comprobar que el vínculo entre una unidad y otra es óptimo.

La sesión de inicio debe direccionar correctamente a la bandeja de entrada, así como la eliminación de un email deberá alojarse de inmediato en la papelera de reciclaje. De existir alguna interfaz errónea se suscitarán problemos en el futuro y los usuarios no podrán hacer uso del sistema.

## ¿Qué errores pueden detectarse con este tipo de pruebas?

Las pruebas de integración de software son esenciales para detectar problemas de conectividad, formato de datos y respuestas inesperadas entre módulos del sistema.

Estas pruebas identifican problemas que surgen cuando los diferentes componentes del software interactúan entre sí.

La pérdida de conectividad, el formateo incorrecto de los datos y las respuestas inesperadas son algunos de los problemas comunes que pueden detectarse mediante las pruebas de integración de software.






La detección oportuna de errores puede minimizar el impacto económico y temporal, de manera que los desarrolladores puedan enfocar esfuerzos en otros proyectos y tareas prioritarias.

## Niveles de pruebas de integración

Existen dos niveles de pruebas de integración: de componentes y de sistemas.

En las pruebas de integración de componentes, se evalúan los elementos integrados en un mismo sistema.

En las pruebas de integración de sistemas, se verifica la relación entre interfaces de sistemas externos.

Las pruebas de integración de componentes se ejecutan antes que las de sistemas, después de verificar las unidades de código individualmente.

Una vez asegurado este paso, es viable realizar los tests que involucran interfaces ajenas al sistema original.

## Tipos de pruebas de integración 


### Big Bang

Concentra todos los módulos de un sistema para comprobar su funcionamiento en conjunto. Viable en proyectos pequeños, pero se pueden pasar por alto errores significativos en proyectos más grandes.

### Ad Hoc

Se plantea una solución para un problema específico y se ejecuta en cualquier momento, recomendado en etapas tempranas para hallar errores no previstos.

### Hybrid (también conocidas como sándwich)

Incluye las prácticas de los modelos top down y down top. El desarrollador puede elegir los módulos posteriores o inferiores simultáneamente con el objetivo de hallar errores en menor tiempo. Requiere de equipo altamente capacitado y es recomendable en sistemas operativos de mayor complejidad.

## Top Down

Las pruebas top down (de arriba hacia abajo) inician el análisis de código en los módulos posteriores, en los que se concentra la información central.

De manera descendiente, conectan con otras interfaces.

En este modelo, pueden detectarse errores de mayor relevancia, y las mejoras pueden implicar más tiempo debido a la complejidad de su composición.

### Down Top

Contrario al modelo anterior, la prueba de integración down top parte de las interfaces inferiores y continúa de manera ascendente. En este caso los problemas son más fáciles de detectar, al
igual que las mejoras a realizar.

Su desventaja es que los módulos complejos se sitúan al final de la prueba y la entrega del producto final puede tomar más tiempo del estimado.
